# Finding Distribution of Genres (Top 20)

```python
name=list(data['Genres'].value_counts()[0:20].keys())
val=list(data['Genres'].value_counts()[0:20])
dict1={}
for i in range(len(name)):
    dict1[name[i]]=val[i]

for j in dict1:
    print(j,':',dict1[j])
```

```python
# Ploting Pie Chart

plt.pie(val,labels=name,autopct="%0.1f%%",radius=3)

plt.show()
```
