```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('googleplaystore.csv')
data.head()
```

```python
data1=pd.read_csv('googleplaystore_user_reviews.csv')
data1
```

```python
d=pd.merge(data,data1,on='App')
d.shape
```

```python
df=d.copy()
df.head(5)
```

# App with Most Positve Sentiments

```python
temp=df[df['Sentiment']=='Positive']
temp['App'].value_counts()[0:5]
name=list(temp['App'].value_counts()[0:5].keys())
val=list(temp['App'].value_counts()[0:5])
```

```python
plt.pie(val,labels=name,autopct='%0.1f%%',radius=1.2)
plt.title('Most Positively Reviewed App')
plt.show()
```

# App with Most Negative Sentiments

```python
temp=df[df['Sentiment']=='Negative']
temp['App'].value_counts()[0:5]
name=list(temp['App'].value_counts()[0:5].keys())
val=list(temp['App'].value_counts()[0:5])
```

```python
plt.pie(val,labels=name,autopct='%0.1f%%',radius=1.2)
plt.title('Most Negatively Reviewed App')
plt.show()
```

# Category Having Highest Average Rating

```python
ans=list(data.groupby('Category')['Rating'].mean()[1:6].keys())
values=list(data.groupby('Category')['Rating'].mean()[1:6])
ans
```

```python
plt.bar(ans,values)
plt.xticks(rotation=45)
plt.show()
```
