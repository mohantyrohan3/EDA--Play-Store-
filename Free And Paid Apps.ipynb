# Average Price of Paid Apps

```python
paid=data[data['Type']=='Paid']
def f(string):
    ans=string[1:]
    return float(ans)
paid['Price']=paid['Price'].apply(f)
paid.head()
```

```python
# Finding Average of Paid Apps

print(paid['Price'].mean())
```
