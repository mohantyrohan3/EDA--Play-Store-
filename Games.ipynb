# Finding Most highly rated games, most installed games , games with most number of reviews 

```python
temp=data[data['Category']=='GAME']
def f(string):
    ans=string[0:len(string)-1]
#     string = string.replace('.', '')
    ans = ans.replace(',', '')
    return float(ans)
temp['Installs']=temp['Installs'].apply(f)
temp.head()
```

```python
# Finding Most installled Game
a=list(temp['Installs'])
arr=np.array(a)
arr1 = np.sort(arr)[::-1]
arr1
temp[temp['Installs']==1.e+09]
```

#### Most Installed Game is Subway Surfers

```python
def f(string):
    return float(string)

temp['Reviews']=temp['Reviews'].apply(f)
ans=temp.sort_values(by='Reviews',ascending=False)
ans.head()
```

#### Most Reviewed Game is Clash Of Clans

```python
ans1=temp.sort_values(by='Rating',ascending=False)
ans2=ans1[0:5]
name=list(ans2['App'])
ins=list(ans2['Installs'])

plt.bar(name,ins,width=0.5)
plt.xticks(rotation=90)
plt.ylabel('No of Installs')
plt.title('Top Rated Games')
plt.show()
```
